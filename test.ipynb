```python
# Read CDC stream
cdc_stream = spark.readStream.format("delta") \
    .option("readChangeFeed", "true") \
    .table("`gds_de_bootcamp_new`.default.raw_upi_transactions")

# Display CDC changes
query = cdc_stream.select(
    "transaction_id",
    "upi_id",
    "merchant_id",
    "transaction_amount",
    "transaction_timestamp",
    "transaction_status",
    "_change_type",  # CDC change type
    "_commit_version",
    "_commit_timestamp"
).writeStream.format("console") \
    .outputMode("append") \
    .start()

query.awaitTermination()
#for testing purposes, you can stop the stream after a certain duration
```
